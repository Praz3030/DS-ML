## Using Dictionary to plot some data

```python
Car_sales={"TATA":81,"FORD":64,"Maruti":56}
fig,ax=plt.subplots(figsize=(5,5))
ax.bar(Car_sales.keys(),Car_sales.values());     #BAR Graph
ax.set(title="Car Sales In India",
      ylabel="Percentage %",
      xlabel="Companies");
```

```python
#Horizontal Bar
Car_sales={"TATA":81,"FORD":64,"Maruti":56}
fig,ax=plt.subplots()
ax.barh(list(Car_sales.keys()),list(Car_sales.values()));
ax.set(title="Car Sales In India",
      xlabel="Percentage %",
      ylabel="Companies");
```

# Heart Disease Data Set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head(10)
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease.head(30).plot(x="age",y="sex",kind="scatter");
```

```python
heart_disease.plot.bar(figsize=(10,30),subplots=True);
```

```python
over_50=heart_disease[heart_disease["age"]>50]
```

```python
over_50.head()
```

```python
# OO Method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
over_50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target',
           figsize=(10,6));
```

```python
# OO from Scratch
fig,ax=plt.subplots(figsize=(10,6))

scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

ax.set(title="Heart Disease Over 50",
      xlabel="Age",
      ylabel="Cholestorol")

ax.legend(*scatter.legend_elements(),title="target")

ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(12,12))
#ax0 customization
ax0.set(title="Analysis 1",
       xlabel="Age",
       ylabel="Cholestrol")
scatter=ax0.scatter(x=over_50["age"],
             y=over_50["chol"],
             c=over_50["target"])
ax0.legend(*scatter.legend_elements(),title='target');
ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

#similar for ax1
ax1.set(title="Analysis 2",
       xlabel="Age",
       ylabel="Cholestrol")
scatter=ax1.scatter(x=over_50["age"],
             y=over_50["chol"],
             c=over_50["fbs"])
ax1.legend(*scatter.legend_elements(),title='fbs');
ax1.axhline(over_50["chol"].mean(),
           linestyle="--");


#figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")
```

```python
over_50["age"].values
```

```python
fig,ax=plt.subplots(figsize=(10,6))
#ax0 customization
ax.hist(over_50['chol'], bins=20, color='skyblue', edgecolor='black');
ax.legend()
plt.show()
```
